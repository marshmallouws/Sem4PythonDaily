```python
# first make sure jsonstat.py is installed: `pip install jsonstat.py`
import jsonstat
#url = 'http://json-stat.org/samples/oecd-canada.json'
url = 'https://api.statbank.dk/v1/data/FOLK1A/JSONSTAT?lang=en&valuePresentation=CodeAndValue&OMR%C3%85DE=000&K%C3%98N=TOT&ALDER=IALT&CIVILSTAND=*'
collection = jsonstat.from_url(url)

# print list of dataset contained into the collection
print(collection)

# select the first dataset of the collection and print a short description
ds = collection.dataset(0)
print(ds)
```

```python
df = ds.to_data_frame()
marital = df[['marital status','Value']]
marital.columns = ['status','no']
print(marital)
marital.plot.bar(x='status',y='no',rot=0)
```
